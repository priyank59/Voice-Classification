```python
X = df.iloc[:,:-1]
X.shape
```

```python
Y = df.iloc[:,-1]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(Y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)
```

```python
X[:10]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
```

```python
svc_model.fit(xtrain,ytrain)
```

```python
y_pred = svc_model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy score is {}%'.format(accuracy_score(ytest,y_pred)*100))
```

```python
print(confusion_matrix(ytest,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001]}
```

```python
grid =GridSearchCV(SVC(),param_grid= param_grid,verbose=2)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid_predictions = grid.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy score is {}%'.format(accuracy_score(grid_predictions,ytest)*100))
print(confusion_matrix(grid_predictions,ytest))
```
